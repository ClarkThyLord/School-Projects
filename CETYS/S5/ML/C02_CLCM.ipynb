```python
x = [x for x in range(1, 11)]
print(x)
x = numpy.linspace(-math.pi, math.pi, 50)
```

```python
y = x

def f1(x, y):
    return math.cos(y) / (1 + x ** 2)
def f2(x, y):
    return f1(x,y) if True else (f1(x, y) - numpy.transpose(f1(x, y))) / 2

z = numpy.zeros((len(x),len(y)))
for i in range(len(x)):
    for n in range(len(y)):
        z[i][n] = f1(x[i], y[n])
plt.contour(x, y, z)
plt.show()
plt.contour(x, y, z, 45)
plt.show()


z = numpy.zeros((len(x),len(y)))
for i in range(len(x)):
    for n in range(len(y)):
        z[i][n] = f2(x[i], y[n])
plt.contour(x, y, z, 15)
plt.show()
```

```python
sns.heatmap(z)
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Plot the surface.
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
