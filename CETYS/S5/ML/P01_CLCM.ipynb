1.9. Realizar un programa que despliegue los siguientes patrones utilizando asteriscos.

```python
size = 0
while True:
    try:
        size = int(input("Tama√±o de Patrones: "))
        if size > 3 and size < 11: break
    except ValueError: pass

result = []
for l in range(size):
    result.append("")
    for c in range(size):
        result[l] += "*" if c - 1 < l else " "

p = ""
for l in result:
    p += l + "\n"
print(p)
p = ""
for l in result:
    p += l[::-1] + "\n"
print(p)

p = ""
result = result[::-1]
for l in result:
    p += l + "\n"
print(p)
p = ""
for l in result:
    p += l[::-1] + "\n"
print(p)
```

1.10. Realizar un programa que despliegue un tri√°ngulo utilizando asteriscos. La altura ser√° definida por el usuario, para el ejemplo de la figura la altura es 5.

```python
size = 0
while True:
    try:
        size = int(input("Tama√±o de Altura: "))
        if size > 0:
            size *= 2
            break
    except ValueError: pass

done = False
result = ""
for l in range(size):
    center = int(size / 2)
    for c in range(size):
        if c == center or c <= center + l and c >= center - l:
            result += "*"
        else:
            result += " "
        
        if c == size - 1:
            if result[-1] == "*":
                done = True
                break
            result += "\n"
    if done: break
print(result)
```
